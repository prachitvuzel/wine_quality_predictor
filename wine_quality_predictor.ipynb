```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df1 = pd.read_csv("C:\\Users\\ADMIN\\Downloads\\archive\\WineQT.csv")
df1_dropped = df1.drop(columns = ['Id'])
df2 = pd.read_csv("C:\\Users\\ADMIN\\Downloads\\winequality-red_improved.csv")
df3= pd.read_csv("C:\\Users\\ADMIN\\Downloads\\winequality-white_improved3.csv")


df5 = pd.concat([df2,df3])


#df5= df4.dropna()
```

```python
columns_to_include = ['fixed acidity','citric acid', 'alcohol','total sulfur dioxide','pH','residual sugar','chlorides','volatile acidity','sulphates']

X = df1[columns_to_include].to_numpy()
Y = df1['quality'].to_numpy()
```

```python
X_train,x_,Y_train,y_=train_test_split(X,Y,test_size= 0.30, random_state =1) 
```

```python
X_cv, X_test, Y_cv, Y_test = train_test_split(x_,y_,test_size= 0.1, random_state =1) 
```

```python
print("Training set:",np.shape(X_train), np.shape(Y_train))
print("Cross-validation set:", np.shape(X_cv), np.shape(Y_cv))
print("Test set:", np.shape(X_test), np.shape(Y_test))
```

```python
scaler_linear = StandardScaler()
X_train_scaled = scaler_linear.fit_transform(X_train)
```

```python
model = Sequential(
   [Dense(units = 400, activation = 'relu',kernel_regularizer = tf.keras.regularizers.L2(0.01)),
    Dense(units = 400, activation = 'relu',kernel_regularizer = tf.keras.regularizers.L2(0.01)),
    Dense(units = 10, activation = 'linear')]
)

model.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits = True),optimizer=  tf.keras.optimizers.Adam(0.01))

model.fit(X_train_scaled, Y_train,epochs = 50)

#,kernel_regularizer = tf.keras.regularizers.L2(0.01)
```

```python

def make_prediction(X):
    yhat_scaled = model.predict(X)
    prediction = tf.nn.softmax(yhat_scaled).numpy()
    row, column = np.shape(yhat_scaled)

    yhat= np.zeros(row)

    for i in range(row):
        yhat[i] = np.argmax(prediction[i])
    return yhat

```

```python
def compute_error(y_pred, y_org):
  correct_predictions = tf.equal(y_pred, y_org)

  accuracy = tf.reduce_mean(tf.cast(correct_predictions, tf.float32))
  error = 1-accuracy

  return error.numpy()
```

```python
X_train_scaled = scaler_linear.fit_transform(X_train)
yhat = make_prediction(X_train_scaled)


error= compute_error(yhat,Y_train)
print(error)
```

```python
X_cv_Scaled = scaler_linear.transform(X_cv)
yhat = make_prediction(X_cv_Scaled)

error= compute_error(yhat,Y_cv)
print(error)
```

```python
X_test_scaled = scaler_linear.transform(X_test)
yhat = make_prediction(X_test_scaled)

error = compute_error(yhat,Y_test)
print(error)
    
```
